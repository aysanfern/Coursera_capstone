# Week 3 Scraping Wikipedia

```python
#importing data
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
type(df)
```

```python
#selecting part of dataframe applicable
df=df[0]
```

```python
#making first row of dataframe our column headers
df=df.rename(columns=df.iloc[0]).drop(df.index[0])
```

```python
df.head()
```

```python
#selecting boroughs that have somethign assigned to their value
norm = df['Borough']!= 'Not assigned'
```

```python
#updating dataframe
df2 = df[norm]
```

```python
df2
```

```python
#ensuring not nulls are in the Neighborhood column
notnull = df2.Neighborhood.notnull()
```

```python
df3 = df2[notnull]
```
